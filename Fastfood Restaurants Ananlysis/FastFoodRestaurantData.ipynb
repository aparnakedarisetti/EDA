```python
restaurant_names = fastfood_restaurant_data['name'].unique()
import nltk
most_similar = []
for i in range(len(restaurant_names)):
    temp_similar = [restaurant_names[i]]
    for j in range(len(restaurant_names)):
        if restaurant_names[i] == restaurant_names[j]:
            continue
        if nltk.edit_distance(restaurant_names[i].lower(), restaurant_names[j].lower()) < 3:
            temp_similar.append(restaurant_names[j])
        if len(temp_similar) > 1:
            most_similar.append(temp_similar)
most_similar_word_count = 0
for i in most_similar:
      most_similar_word_count += len(i)
print("size of the most similar list:", most_similar_word_count)
```

```python
def sortFirst(val): 
    return val[0]  

most_similar.sort(key = sortFirst)  
most_similar
unique_values = [list(x) for x in set(tuple(x) for x in most_similar)]
unique_values
```

```python
def percent_more_than_1_restaurants_city(list_restaurants,city):
    more_than_1_restaurants = list_restaurants[list_restaurants > 1]
    percent_more_than_1_restaurants = round(len(more_than_1_restaurants)/sum(list_restaurants)*100,2)
    print("City "+city+ " has "+ str(percent_more_than_1_restaurants)+ "% of total fastfood restaurants more than 1 restaurant open.")
          
percent_more_than_1_restaurants_city(restaurants, "Cincinnati")
percent_more_than_1_restaurants_city(restaurants1, "Las Vegas")
percent_more_than_1_restaurants_city(restaurants2, "Houston")
```

```python
def plot_donut_chart(city_name):
    data = fastfood_restaurant_data[fastfood_restaurant_data["city"]== city_name]
    counts =  data["names"].value_counts()[:10]
    space = np.ones(10)/10
    counts.plot(kind='pie', explode=space, fontsize=14, autopct='%3.1f%%', wedgeprops=dict(width=0.15), 
                                       shadow=True, startangle=160, figsize=(10,10), cmap='inferno', legend=True)
    plt.ylabel('')
    plt.title("Spread of fastfood restaurants in "+ city_name,fontsize=18,fontweight="bold") 
    plt.show()

plot_donut_chart("Cincinnati")
plot_donut_chart("Las Vegas")
plot_donut_chart("Houston")
```

#### 4.1.0 Geographical Representation of fastfood restaurants spread across provinces in US

```python
states = fastfood_restaurant_data['province'].value_counts().index.tolist()
counts_of_restaurants_by_states = fastfood_restaurant_data['province'].value_counts()

data= [dict(type='choropleth',
            locations = states, # Spatial coordinates
            z = counts_of_restaurants_by_states, # Data to be color-coded
            locationmode = 'USA-states', # set of locations match entries in `locations`
            colorscale = 'Blues',
            marker_line_color = 'white',
            colorbar_title = "Number of Fast Fast Restaurants"
        )]
layout = dict(title = 'Fast Food Restaurants by Province in US', 
              geo = dict(scope='usa'))
iplot(dict(data=data, layout=layout))
```

**Observation:** From the above plot, we see that Calfornia,Texas and Ohio are in dark blue color which represents more number of fast food restaurants and the provinces on the top right corner are in white color which indicate that there are no fast food restaurants.

#### 4.1.2 Fast food restaurants vs Population

We have to import the population data in US to do further analysis.

di_population = {}

    di_population[population_data.values[i][0]] = population_data.values[i][1]
di_population

```python
province_counts = fastfood_restaurant_data['province'].value_counts()
province_counts.head()
```

```python
us_province_names = pd.read_csv('US-2-letter-state-code.csv')
us_province_names.head()
```

```python
di_province = {}
for i in range(len(us_province_names)):
    di_province[us_province_names.values[i][1]] = us_province_names.values[i][0]
```

```python
def convert_state_names(state_code):
    return di_province[state_code]

fastfood_restaurant_data['province_full'] = ""
for i in range(len(fastfood_restaurant_data['province'])):
    fastfood_restaurant_data.loc[i,'province_full'] = convert_state_names(fastfood_restaurant_data.loc[i,'province']) 
    
```

```python
counts_province = fastfood_restaurant_data['province_full'].value_counts()
```

#### 4.1.3 What is the restaurant per person ratio for each province?

```python
per_capita_count = [None] * len(counts_province)
provinces = list(counts_province.index)
for i in range(len(counts_province)):
    per_capita_count[i] = counts_province[provinces[i]] / int(di_population[provinces[i]].replace(",",""))
per_capita_count[:10]
```

```python
data= [dict(type='choropleth',
            locations = list(di_province.keys()), 
            z = per_capita_count, 
            locationmode = 'USA-states',
            colorscale = 'reds',
            marker_line_color = 'white',
            colorbar_title = "Restaurants opened / population"
        )]
layout = dict(title = 'Ratio between number of fast food Restaurants and population', 
              geo = dict(scope='usa'))
iplot(dict(data=data, layout=layout))
```

**Observation:** From the above plot, we observe that the fast food restaurants are opened based on the population in each province which makes sense because restaurants need customers to visit their restaurants.
