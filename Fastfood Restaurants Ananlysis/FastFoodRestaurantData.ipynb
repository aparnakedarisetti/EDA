Change the restaurant names uniformly over the data.

```python
fastfood_restaurant_data['name'].nunique()
```

```python
restaurant_names = fastfood_restaurant_data['name'].unique()
import nltk
most_similar = []
for i in range(len(restaurant_names)):
    temp_similar = [restaurant_names[i]]
    for j in range(len(restaurant_names)):
        if restaurant_names[i] == restaurant_names[j]:
            continue
        if nltk.edit_distance(restaurant_names[i].lower(), restaurant_names[j].lower()) < 3:
            temp_similar.append(restaurant_names[j])
        if len(temp_similar) > 1:
            most_similar.append(temp_similar)
most_similar_word_count = 0
for i in most_similar:
      most_similar_word_count += len(i)
print("size of the most similar list:", most_similar_word_count)
most_similar
```

```python
def sortFirst(val): 
    return val[0]  

most_similar.sort(key = sortFirst)  
most_similar
unique_values = [list(x) for x in set(tuple(x) for x in most_similar)]
unique_values
```

```python
def percent_more_than_1_restaurants_city(list_restaurants,city):
    more_than_1_restaurants = list_restaurants[list_restaurants > 1]
    percent_more_than_1_restaurants = round(len(more_than_1_restaurants)/sum(list_restaurants)*100,2)
    print("City "+city+ " has "+ str(percent_more_than_1_restaurants)+ "% of total fastfood restaurants more than 1 restaurant open.")
          
percent_more_than_1_restaurants_city(restaurants, "Cincinnati")
percent_more_than_1_restaurants_city(restaurants1, "Las Vegas")
percent_more_than_1_restaurants_city(restaurants2, "Houston")
```

#### 4.4 Top 10 fastfood restaurants spread across Cincinnati,Las Vegas and Houstan:

```python
def plot_pie_chart(city_name):
    counts =  fastfood_restaurant_data[fastfood_restaurant_data["city"]==city_name]["names"].value_counts()[:10].values
    labels = fastfood_restaurant_data[fastfood_restaurant_data["city"]==city_name]["names"].value_counts().index.tolist()[:10]
    plt.figure(figsize=(22,5))
    plt.plot(labels, counts, color='green', linestyle='dashed', linewidth = 3, 
             marker='o', markerfacecolor='blue', markersize=12) 
    plt.xlabel('Fastfood restaurants',fontsize=15) 
    plt.ylabel('Number of fastfood restaurants',fontsize=15) 
    plt.title("Spread of fastfood restaurants in "+ city_name,fontsize=18,fontweight="bold") 
    plt.show() 

plot_pie_chart("Cincinnati")
plot_pie_chart("Las Vegas")
plot_pie_chart("Houston")
```

#### 4.6 Geographical Representation of fastfood restaurants spread across provinces in US
